## Number of Creatures By Type

```python
# data
sf = pd.read_json("https://c2.scryfall.com/file/scryfall-bulk/oracle-cards/oracle-cards-20220615090211.json")
```

```python
sf.columns
```

```python
# tidy scryfall data

## general clean up

### reduce columns
sf_cols = ['name', 'mana_cost', 'cmc', 'type_line', 'oracle_text', 'colors', 'color_identity', 'edhrec_rank', 'power', 'toughness']
sf2 = sf[sf_cols]

### change data types
sf2['color_identity'] = sf2['colors'].astype(str)

## filter for just humans for volo
humans = sf2[
    (sf2['type_line'].str.contains('Human')) & 
    (~sf2['type_line'].str.contains('//', regex = False)) &
    (~sf2['color_identity'].str.contains('(B|R|G)', regex = True)) & 
    (~sf2['mana_cost'].isna())
]

humans[['type', 'subtype']] = humans['type_line'].str.split('—', expand = True)
humans_exp = humans.assign(subtype = humans['subtype'].str.split(' ')).explode('subtype')
humans_exp = humans_exp[(humans_exp['subtype'] != 'Human') & (humans_exp['subtype'] != '')]

## filter for just wizards for volo
wizards = sf2[
    (sf2['type_line'].str.contains('Wizard')) & 
    (~sf2['type_line'].str.contains('//', regex = False)) &
    (~sf2['color_identity'].str.contains('(B|R|G)', regex = True)) & 
    (~sf2['mana_cost'].isna())
]

wizards[['type', 'subtype']] = wizards['type_line'].str.split('—', expand = True)
wizards_exp = wizards.assign(subtype = wizards['subtype'].str.split(' ')).explode('subtype')
wizards_exp = wizards_exp[(wizards_exp['subtype'] != 'Wizard') & (wizards_exp['subtype'] != '')]
```

```python
# count human creature representations
humans_exp.groupby('subtype').size().reset_index(name = 'count')\
    .sort_values(by = 'count', ascending = False)
```

```python
# count wozard creature representations
wizards_exp.groupby('subtype').size().reset_index(name = 'count')\
    .sort_values(by = 'count', ascending = False)
```
